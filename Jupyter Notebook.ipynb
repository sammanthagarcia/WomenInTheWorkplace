Gender inequality and racial discrimination in education and in the workplace are most prevalent among Women. As a result, this significantly affects one's level of education, income, and occupation. In this Jupyter Notebook, I plan to explore how race plays a role in highest level of education acheived and income. Using datasets from the United States Census Bureau surveyed in 2019, I will use horizontal bar charts to determine the correlation between different factors.  

```python
# Import Python libraries necessary for analyzing and manipulating data
import math 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Import csv files
women25_to_64_White_alone = pd.read_csv('Women 25-64 (White Alone).csv')
# BAH is an acronym for Black, Asian, and Hispanic
women25_to_64_BAH = pd.read_csv('Women 25-64 (Black, Asian, Hispanic).csv')
```

```python
# Data cleanup 
# In this cell, I removed rows that would cause an error in my graphs 
#   (such as, Median Earnings, Mean Earnings, Standard Error, and Gini Ratio)
women25_to_64_White_alone = women25_to_64_White_alone.drop(['Total', 'Total.1'], axis=1).drop([0, 1, 2]).drop(range(44, 50))
df_White = pd.DataFrame(women25_to_64_White_alone)

# In Google Sheets, I summed data from three Races: Black, Asian, and Hispanic. 
    #In the following code, I used .iloc from pandas to select the 
    # columns and rows that represent the sum of columns. 
women25_to_64_BAH = women25_to_64_BAH.iloc[52: 96, 14:26].drop(['Total', 'Total.1'], axis=1).drop([52, 53, 54])
df_BAH = pd.DataFrame(women25_to_64_BAH)
```

In the following cells, I  divided the rows (income brackets) into groups of 7, so the visualizations are easier to read. I also removed commas (",") from the tables so I can convert the data to integers.

### Conclusion
White Women and Women of Color (Black, Asian, Hispanic) have similar occupational categories, however, there are more White Women employed in the workforce. In addition, looking at how the occupations are ranked for each group, Women of Color are more likely to hold jobs in "service occupations." From the Bureau of Labor Statistics, service occupations consist of: 
 - Animal Care and Service Workers 
 - Barbers, Hairstylists, and Cosmetologists 
 - Childcare Workers 
 - Fitness Trainers and Instructors 
 - Funeral Service Workers 
 - Gambling Services Workers 
 - Manicurists and Pedicurists 
 - Recreation Workers 
 - Skincare Specialists 
 
 In the same article, these service jobs only require a High School Diploma (or equivalent), Postsecondary Nondegree Award, or an Associate's Degree. 
 
On the other hand, White Women are more likely to hold positions in management and professional related occupations. 

With this information from the Bureau of Labor Statistics, we know that Women of Color have these specific occupations and earn a lower educational level - this supports the fact that Women of Color are still facing discrimination.
